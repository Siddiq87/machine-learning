```python
import numpy as np
```

```python
df["age_group"] = np.where(df.age > 60, "s1", "s2")
df
```

```python
def find_age_group(age):
    if age > 60:
        return "s1"
    elif age > 40:
        return "s2"
    else:
        return "s3"
```

```python
df["age_group"] = df.age.apply(find_age_group)
df
```
