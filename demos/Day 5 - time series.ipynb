```python
y_log = np.log10(y)
```

```python
y_log.plot()
```

```python
y_log.describe()
```

```python
y_log = y_log.asfreq("B", method="ffill")
y_log
```

```python
result = pd.DataFrame({
    "actual": y_log, 
    "rolling_mean": y_log.rolling(5).mean(),
    "rolling_std": y_log.rolling(5).std(),
    "diff1": y_log.diff(1),
    "ewm": y_log.ewm(3).mean()
})

result["ma_lag1"] = result["rolling_mean"].shift(2)
result
```

```python
result_dropped = result.dropna()
result_dropped.head()
```

```python
metrics.mean_squared_error(result_dropped.actual, result_dropped.ma_lag1) ** 0.5
```

```python
metrics.mean_squared_error(result_dropped.actual, result_dropped["ewm"]) ** 0.5
```

```python
result[["actual", "rolling_mean"]].plot()
```

```python
result["jerks"] = np.abs(result.actual - result.rolling_mean) > 3 * result.rolling_std
```

```python
result[result.jerks]
```
