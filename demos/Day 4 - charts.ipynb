```python
import numpy as np
```

```python
def find_outlier(y):
    q1, q3 = np.percentile(y, [25, 75])
    iqr = q3-q1
    uw = min(y.max(), q3 + 1.5 * iqr)
    lw = max(y.min(), q1 - 1.5 * iqr)
    print("upper whisker: ", uw, "lower whisker: ", lw)
    return (y>uw) | (y<lw)


    
```

```python
df[find_outlier(df.charges)]
```
