```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("/Users/abasar/Downloads/maruti.csv")
```

```python
df.head()
```

```python
df.index = pd.to_datetime(df["Date"])
```

```python
df.head()
```

```python
df[df.index>"2019-09"]
```

```python
price = df[["Average Price"]]
price.columns =["y"]
price["y_est"] = np.nan
price["y_std"] = np.nan
price.y = np.log(price.y)

price.y_std = price.y.rolling(3).std()
price.head(10)
```

```python
price.head(10)
```

```python
price.y.plot()
plt.hlines([np.mean(price.y)], price.index.min(), price.index.max(), linestyles="--", lw = 1)
```

```python
price["y_est"] = price.y.rolling(3).mean().shift(1)
```

```python
price.head(10)
```

```python
price["d1"] = price.y.diff(1)
```

```python
price.head()
```

```python
price.d1.plot()
```

```python
price.d1.cumsum().plot()
```

```python
from sklearn import metrics
```

```python
tmp = price.dropna()
```

```python
metrics.r2_score(tmp.y, tmp.y_est)
```

```python
price["d1_est"] = price["d1"].rolling(3).mean().shift(1)
```

```python
price.head(10)
```

```python
tmp = price.dropna()
metrics.r2_score(tmp.d1, tmp.d1_est)
```
