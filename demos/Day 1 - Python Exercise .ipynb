```python
count_by_symbol = {}

for line in lines:
    parts = line.split(",")
    symbol = parts[-1]
    count = count_by_symbol.get(symbol, 0)
    count_by_symbol[symbol] = count + 1
    
print(count_by_symbol)
    
```

```python
counts = list(count_by_symbol.items())
```

```python
sorted(counts, key=lambda t: t[1], reverse=True)[:10]
```

```python
def find_count(t):
    return t[1]
```

```python
sorted(counts, key=find_count, reverse=True)[:10]
```

```python
count_by_symbol = {}
sum_by_symbol = {}

for line in lines:
    parts = line.split(",")
    symbol = parts[-1]
    volume = float(parts[5])
    count = count_by_symbol.get(symbol, 0)
    count_by_symbol[symbol] = count + 1
    current_sum = sum_by_symbol.get(symbol, 0)
    sum_by_symbol[symbol] = current_sum + volume
    
print(count_by_symbol, sum_by_symbol)
```

```python
avg_by_sym = {}
for symbol in count_by_symbol:
    avg_by_sym[symbol] = sum_by_symbol[symbol]/count_by_symbol[symbol]
```
